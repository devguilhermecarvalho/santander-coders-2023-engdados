```python
#Exércício 04
estoque = {}

def adicionar_produto():
    nome = input("Digite o nome do produto: ")
    preco = float(input("Digite o preço do produto: "))
    quantidade = int(input("Digite a quantidade em estoque: "))
    
    estoque[nome] = {"preco": preco, "quantidade": quantidade}
    print("Produto adicionado ao estoque.")

def atualizar_preco():
    nome = input("Digite o nome do produto: ")
    
    if nome in estoque:
        novo_preco = float(input("Digite o novo preço do produto: "))
        estoque[nome]["preco"] = novo_preco
        print("Preço atualizado com sucesso.")
    else:
        print("Produto não encontrado.")

def vender_produto():
    nome = input("Digite o nome do produto: ")
    
    if nome in estoque:
        quantidade_vendida = int(input("Digite o número de unidades vendidas: "))
        estoque[nome]["quantidade"] -= quantidade_vendida
        print("Venda registrada.")
    else:
        print("Produto não encontrado.")

def imprimir_estoque():
    print("Lista de Produtos:")
    for produto, info in estoque.items():
        nome = produto
        preco = info["preco"]
        quantidade = info["quantidade"]
        print(f"Nome: {nome} | Preço: R${preco:.2f} | Quantidade: {quantidade}")

# Menu de opções
while True:
    print("\n==== Gerenciamento de Estoque ====")
    print("1 - Adicionar produto ao estoque")
    print("2 - Atualizar preço de um produto")
    print("3 - Vender um produto")
    print("4 - Imprimir estoque")
    print("0 - Sair")

    opcao = input("Digite a opção desejada: ")

    if opcao == "1":
        adicionar_produto()
    elif opcao == "2":
        atualizar_preco()
    elif opcao == "3":
        vender_produto()
    elif opcao == "4":
        imprimir_estoque()
    elif opcao == "0":
        print("Encerrando o programa...")
        break
    else:
        print("Opção inválida. Digite uma opção válida.")
```

```python
#Exercício 05
eleicao = {}

def registrar_voto():
    candidato = input("Digite o nome do candidato: ")
    
    if candidato in eleicao:
        eleicao[candidato] += 1
        print(f"Voto registrado para o candidato {candidato}.")
    else:
        print("Candidato não encontrado.")

def verificar_votos_candidato():
    candidato = input("Digite o nome do candidato: ")
    
    if candidato in eleicao:
        total_votos = eleicao[candidato]
        print(f"O candidato {candidato} tem {total_votos} votos.")
    else:
        print("Candidato não encontrado.")

def verificar_total_votos():
    total_votos = sum(eleicao.values())
    print(f"O número total de votos contabilizados é {total_votos}.")

def imprimir_resultados():
    print("Resultados da Eleição:")
    for candidato, votos in eleicao.items():
        print(f"Candidato: {candidato} | Votos: {votos}")

# Menu de opções
while True:
    print("\n==== Sistema de Votação ====")
    print("1 - Registrar voto para um candidato")
    print("2 - Verificar número total de votos de um candidato")
    print("3 - Verificar número total de votos contabilizados")
    print("4 - Imprimir resultados da eleição")
    print("0 - Encerrar a eleição")

    opcao = input("Digite a opção desejada: ")

    if opcao == "1":
        registrar_voto()
    elif opcao == "2":
        verificar_votos_candidato()
    elif opcao == "3":
        verificar_total_votos()
    elif opcao == "4":
        imprimir_resultados()
    elif opcao == "0":
        print("Encerrando a eleição...")
        break
    else:
        print("Opção inválida. Digite uma opção válida.")
```
