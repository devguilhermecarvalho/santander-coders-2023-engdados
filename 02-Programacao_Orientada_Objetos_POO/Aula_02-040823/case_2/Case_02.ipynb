Case 2: Como representar um cliente de banco e uma conta bancária em Python, onde toda conta possui um cliente e vice-versa.


```python
import json

def lerArquivo():
    path = 'dados_clientes.json'
    with open(path, 'r', encoding='utf-8') as arquivo:
        try:
            dados = json.load(arquivo)
            print("Dados lidos com sucesso.")
            return dados
        except json.JSONDecodeError:
            print("Erro de leitura.")
            return {}

dados = lerArquivo()
print(f"{dados}")
```

```python
class Clientes:
    def __init__(self):
        self.clientes = dados

    def inserir_cliente(self):
        novo_cliente = {
            "Id": input(""),
            "Nome": input("Nome: "),
            "CPF": input("CPF: "),
            "Status": True
        }
        
        while not novo_cliente["Id"].strip():
            novo_cliente["Id"] = input("Digite o ID do Cliente:")
            if not novo_cliente["Id"].strip():
                print("O Id não pode ser vazio.")

        while not novo_cliente["Nome"].strip():
            novo_cliente["Nome"] = input("Digite o Nome do Cliente:")
            if not novo_cliente["Nome"].strip():
                print("O nome não pode ser vazio.")

        while not novo_cliente["CPF"].strip():
            novo_cliente["CPF"] = input("Digite o CPF do Cliente:")
            if not novo_cliente["CPF"].strip():
                print("O CPF não pode ser vazio.")
        
        print(f"{novo_cliente}\n")
        print("\n\tAs informações do usuário estão corretas?")
        resposta = input("\tDigite 's' para sim ou 'n' para não: ")
        if resposta.lower() == 's':
            chave_maxima = str(int(max(self.clientes.keys(), default=0)) + 1)
            self.clientes[chave_maxima] = novo_cliente
            print("\tInformações salvas. O ID do novo usuário é", chave_maxima)
        else:
            print("\tAs informações não foram salvas.")
        return novo_cliente 
        

    def listar_registros_clientes(self):
        for chave, registro in self.clientes.items():
            if registro.get("Status") is True:
                print(f"Registro: {chave}:{registro}")

    def atualiza_registros_clientes(self):
        pass

clientes_obj = Clientes()
clientes_obj.inserir_cliente()
clientes_obj.listar_registros_clientes()
```

```python
class ContaCorrente:
    def __init__(self, clientes_obj):
        self.clientes_obj = clientes_obj

    def adiciona_conta_corrente(self, id, conta_corrente):
        if id in self.clientes_obj.clientes:
            self.clientes_obj.clientes[id]["conta_corrente"] = conta_corrente
            print(f"Conta Corrente adicionada para o cliente {id}.")
        else:
            print(f"Cliente id {id} não encontrado.")
```

```python
clientes_obj = Clientes()
clientes_obj.inserir_cliente()
clientes_obj.listar_registros_clientes()

```

```python
gerenciador_contas = ContaCorrente(clientes_obj)

gerenciador_contas.adiciona_conta_corrente("4", "7194")

# Imprimir os registros atualizados
for id, registro in clientes_obj.clientes.items():
    print(f"Registro {id}: {registro}")
```

```python
def atualiza_arquivo(dados):
    path = 'dados_clientes.json'
    with open(path, 'w') as arquivo:
        json.dump(dados, arquivo, indent=4)

atualiza_arquivo(dados)
```
